## function

```python
def add(x,y):
    c=x+y
    return c
def sub(x,y):
    d=x-y
    return d
def mul(x,y):
    e=x*y
    return e
def division(x,y):
    f=x/y
    return f
a=int(input())
b=int(input())
print(add(a,b))
print(sub(a,b))
print(mul(a,b))
print(division(a,b))
```

# even or odd

```python
def even_or_odd(x):
    if x%2==0:
        print("it is a even num")
    else:
        print("it is odd num")
n=int(input("enter a number")) 
even_or_odd(n)
```
